```python
# Importing all the necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/heart.csv")   # Importing the data
```

```python
data
```

Splitting the data into IV and DV

```python
X = data.drop('output',axis = 1)
```

```python
X
```

```python
y = data["output"]
```

```python
y
```

Training the ANN model

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
xtrain.shape
```

```python
xtest.shape
```

Creating ANN model architecture

```python
model = Sequential() #BRAIN WITHOUT NEURON
```

```python
model.add(Dense(100, activation="relu",input_shape = (13,))) #input layer
```

```python
model.add(Dense(200, activation="relu")) #hidden layer
model.add(Dense(300, activation="relu")) #hidden layer
model.add(Dense(200, activation="relu")) #hidden layer
model.add(Dense(400, activation="relu")) #hidden layer
```

```python
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer="adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(xtrain,ytrain,epochs = 150, batch_size=15)
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
ypred_transformed = []
for i in ypred:
    if i >= 0.5:
        ypred_transformed.append([1])
    else:
        ypred_transformed.append([0])
```

```python
ypred_transformed
```

```python
import numpy as np
ypred_transformed = np.array(ypred_transformed)
```

```python
ypred_transformed
```

```python
accuracy_score(ypred_transformed,ytest)
```

```python
confusion_matrix(ypred_transformed,ytest)
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier
import numpy as np

# Assuming xtrain and ytrain are your training data
# Perform any necessary preprocessing here (e.g., normalization or scaling)
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)

# Function to build the Keras model
def build_classifier():
    model = Sequential()
    model.add(Dense(100, activation="relu", input_shape=(13,)))
    model.add(Dense(200, activation="relu"))  # hidden layer
    model.add(Dense(300, activation="relu"))  # hidden layer
    model.add(Dense(200, activation="relu"))  # hidden layer
    model.add(Dense(400, activation="relu"))  # hidden layer
    model.add(Dense(1, activation="sigmoid"))
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return model

# Wrap the Keras model in a scikit-learn compatible wrapper
model = KerasClassifier(build_fn=build_classifier, batch_size=15, epochs=100, verbose=0)

# Use StratifiedKFold for cross-validation
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# Use cross_val_score with the defined model, kfold, and your data
accuracies = cross_val_score(estimator=model, X=xtrain, y=ytrain, cv=kfold, n_jobs=-1, scoring='accuracy')

# Print the accuracies or use them as needed
print("Accuracies:", accuracies)
print("Mean Accuracy:", accuracies.mean())
```

Hyper-parameter tuning

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Assuming xtrain and ytrain are your training data
# Perform any necessary preprocessing here (e.g., normalization or scaling)
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)

# Function to build the Keras model
def build_classifier(optimizer = 'adam'):
    model = Sequential()
    model.add(Dense(100, activation="relu", input_shape=(13,)))
    model.add(Dense(200, activation="relu"))  # hidden layer
    model.add(Dense(300, activation="relu"))  # hidden layer
    model.add(Dense(200, activation="relu"))  # hidden layer
    model.add(Dense(400, activation="relu"))  # hidden layer
    model.add(Dense(1, activation="sigmoid"))
    model.compile(optimizer= optimizer, loss="binary_crossentropy", metrics=["accuracy"])
    return model

# Wrap the Keras model in a scikit-learn compatible wrapper
model = KerasClassifier(build_fn=build_classifier)
parameters = {'batch_size':[10,15,20],'epochs':[100,200],'optimizer':['adam','rmsprop']}
# 'cv':[3,5,10]

grid_search = GridSearchCV(estimator = model, param_grid = parameters, scoring = 'accuracy',cv = 10)

grid_search = grid_search.fit(xtrain,ytrain)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
