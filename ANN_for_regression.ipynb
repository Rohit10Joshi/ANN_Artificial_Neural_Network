```python
# Importing all the necessary libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv("/content/expenses.csv")   # Importing the data
```

```python
data
```

Data Preprocessing

```python
le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
data['smoker'] = le.fit_transform(data['smoker'])
data['region'] = le.fit_transform(data['region'])
```

```python
data
```

Splitting the data into IV and DV

```python
X = data.drop('charges', axis = 1).values
```

```python
X
```

```python
y = data['charges'].values
```

```python
X.shape
```

```python
type((6,)) # input shape
```

Training the ANN Regression model

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2,random_state=50)

model = Sequential() #brain without neurons

model.add(Dense(10,activation = "relu", input_shape = (6,))) # input layer

model.add(Dense(20,activation = "relu")) # hidden layer
model.add(Dense(30,activation = "relu")) # hidden layer
model.add(Dense(50,activation = "relu")) # hidden layer
model.add(Dense(10,activation = "relu")) # hidden layer

model.add(Dense(1)) #output layer neuron

model.compile(optimizer = "adam", loss = "mean_squared_error")
```

```python
batch_size = 32
```

```python
model.fit(xtrain,ytrain, epochs = 100, batch_size  = batch_size)
```

```python
ypred = model.predict(xtest)
```

```python
r2_score = r2_score(ytest,ypred)  # Checking correctness of model
print(f'r2_score value is- {r2_score:.2f}')
```

Cross Validating the data

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Create an ANN regressor model
ann_regressor = MLPRegressor(hidden_layer_sizes=(100,200,300,400,500), max_iter=500)

# Set up k-fold cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation and calculate mean squared error and R2 score for each fold
r2_scores = cross_val_score(ann_regressor, X, y, scoring='r2', cv=kfold)
```

```python
# Print the mean and standard deviation of MSE and R2 scores
print(f'Mean R2 Score: {np.mean(r2_scores):.2f}')
```
